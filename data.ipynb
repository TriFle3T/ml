```python
def most_frequent1(data):
    emotion_label = {'Happiness':0, 'Angry':1, 'Disgust':2, 'Fear':3, 'Neutral':4, 'Sadness':5, 'Surprise':6}
    emotion={'Happiness':0, 'Angry':0, 'Disgust':0, 'Fear':0, 'Neutral':0, 'Sadness':0, 'Surprise':0}
    for i in range(5):
        emotion[data[2*i]] += data[2*i+1] if data[2*i+1] != 0 else 1
    return emotion_label[max(emotion,key=emotion.get)]

def most_frequent2(data):
    emotion_label = {'happiness':0, 'angry':1, 'disgust':2, 'fear':3, 'neutral':4, 'sadness':5, 'surprise':6}
    emotion={'happiness':0, 'angry':0, 'disgust':0, 'fear':0, 'neutral':0, 'sadness':0, 'surprise':0}
    for i in range(5):
        emotion[data[2*i]] += data[2*i+1] if data[2*i+1] != 0 else 1
    return emotion_label[max(emotion,key=emotion.get)]
```

```python
data1 = data1[data1['발화문'].str.contains('취업')== False]
data2 = data2[data2['발화문'].str.contains('취업')== False]
data3 = data3[data3['발화문'].str.contains('취업')== False]
data4 = data4[data4['Sentence'].str.contains('취업')== False]
```

```python
# data1.loc[(data1['상황'] == "anger"), '상황'] = 1  #분노 => 1
# data1.loc[(data1['상황'] == "disgust"), '상황'] = 2  #혐오 => 2
# data1.loc[(data1['상황'] == "fear"), '상황'] = 3  #공포 => 3
# data1.loc[(data1['상황'] == "neutral"), '상황'] = 4  #중립 => 4
# data1.loc[(data1['상황'] == "sad"), '상황'] = 5  #슬픔 => 5

# data2.loc[(data2['상황'] == "anger"), '상황'] = 1  #분노 => 1
# data2.loc[(data2['상황'] == "disgust"), '상황'] = 2  #혐오 => 2
# data2.loc[(data2['상황'] == "fear"), '상황'] = 3  #공포 => 3
# data2.loc[(data2['상황'] == "neutral"), '상황'] = 4  #중립 => 4
# data2.loc[(data2['상황'] == "sad"), '상황'] = 5  #슬픔 => 5

# data3.loc[(data3['상황'] == "happiness"), '상황'] = 0  #행복 => 0
# data3.loc[(data3['상황'] == "angry"), '상황'] = 1  #분노 => 1
# data3.loc[(data3['상황'] == "disgust"), '상황'] = 2  #혐오 => 2
# data3.loc[(data3['상황'] == "fear"), '상황'] = 3  #공포 => 3
# data3.loc[(data3['상황'] == "neutral"), '상황'] = 4  #중립 => 4
# data3.loc[(data3['상황'] == "sadness"), '상황'] = 5  #슬픔 => 5
# data3.loc[(data3['상황'] == "surprise"), '상황'] = 6  #놀람 => 6

data4.loc[(data4['Emotion'] == "행복"), 'Emotion'] = 0  #행복 => 0
data4.loc[(data4['Emotion'] == "분노"), 'Emotion'] = 1  #분노 => 1
data4.loc[(data4['Emotion'] == "혐오"), 'Emotion'] = 2  #혐오 => 2
data4.loc[(data4['Emotion'] == "공포"), 'Emotion'] = 3  #공포 => 3
data4.loc[(data4['Emotion'] == "중립"), 'Emotion'] = 4  #중립 => 4
data4.loc[(data4['Emotion'] == "슬픔"), 'Emotion'] = 5  #슬픔 => 5
data4.loc[(data4['Emotion'] == "놀람"), 'Emotion'] = 6  #놀람 => 6

data1_values=[]
for data in data1.loc[:,"1번 감정":"5번 감정세기"].values.tolist():
    data1_values.append(most_frequent1(data))

data2_values=[]
for data in data2.loc[:,"1번 감정":"5번 감정세기"].values.tolist():
    data2_values.append(most_frequent1(data))

data3_values=[]
for data in data3.loc[:,"1번 감정":"5번 감정세기"].values.tolist():
    data3_values.append(most_frequent2(data))

data_list = []
for ques, label in zip(data1['발화문'], data1_values)  :
    data1_element = []   
    data1_element.append(ques)
    data1_element.append(str(label))

    data_list.append(data1_element)

for ques, label in zip(data2['발화문'], data2_values)  :
    data2_element = []   
    data2_element.append(ques)
    data2_element.append(str(label))

    data_list.append(data2_element)

for ques, label in zip(data3['발화문'], data3_values)  :
    data3_element = []   
    data3_element.append(ques)
    data3_element.append(str(label))

    data_list.append(data3_element)

for ques, label in zip(data4['Sentence'], data4['Emotion'])  :
    data4_element = []   
    data4_element.append(ques)
    data4_element.append(str(label))

    data_list.append(data4_element)
```

```python
from collections import Counter
Counter(datatest['label'])
```

```python
from imblearn.under_sampling import RandomUnderSampler
def ratio_multiplier(y):
    multiplier = {1:0.5,4:0.6,5:0.27}
    target_stats = Counter(y)
    for key, value in target_stats.items():
        if key in multiplier:
            target_stats[key] = int(value * multiplier[key])
    return target_stats

undersample = RandomUnderSampler(random_state=0,sampling_strategy=ratio_multiplier)
X_resampled, y_resampled = undersample.fit_resample(np.reshape(datatest['ques'].tolist(),(-1,1)),datatest['label'])
```

```python
data8_list=[]
for ques, label in zip(X_resampled[:,0].tolist(), y_resampled.tolist())  :
    data8_element = []   
    data8_element.append(ques)
    data8_element.append(str(label))

    data8_list.append(data8_element)
```

```python
data8 = pd.DataFrame(data8_list,columns=['ques','label'])
```
